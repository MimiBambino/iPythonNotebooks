#Coffee Cooling Lab

###I determine whether one should put cream in coffee at the coffee shop or back at the office in order to keep the coffee as hot as possible.

First we set our global variables for the outside temperature and the r value, which is a cooling coefficient dependent on things like cup material and air flow.  I just made these numbers up.

```python
temp_out = 70.0
r = 0.9
```

###Now we define our function based on Newton's cooling law:

```python
def cooling_law(temp_start, walk_time):
    return temp_out + (temp_start - temp_out) * r ** walk_time
```

###Let's call the function assuming an intial temperature for the coffee of 185 degrees fahrenreit and a walk time of ten minutes.

```python
cooling_law(185.0, 10.0)
```

###Wouldn't it be more interesting to see how the coffee cools as we walk?
###Here we call the function, passing in a numpy array as the second parameter.

```python
times = arange(0.,21.,1.)
temperatures = cooling_law(185., times)
temperatures
```

###This returns an array of the temperature over time, every minute from 0 to 20.
###Let's plot it!

```python
plot(times, temperatures, 'o')
```

###So now we know how the temperature would change if we never added cream.  But I like cream, so how does this help me?
###We need a new function that will account for the cooling effect of the cream.

```python
def temp_mixture(t1, v1, t2, v2):
    '''
    return the temperature of coffee after mixing it with cream.
    t1 is the temperature of coffee
    v2 is the volume of coffee
    t2 is the temperature of cream
    v2 is the volume of cream
    '''
    return (t1 * v1 + t2 * v2) / (v1 + v2)
```

```python
temp_coffee = 185.0
temp_cream = 40.0
vol_coffee = 8.0
vol_cream = 1.0
```

###Now let's find out how the coffee cools if we add cream at the coffee shop.

```python
initial_temp_mix_at_shop = temp_mixture(temp_coffee, vol_coffee, temp_cream, vol_cream)
temperatures_mix_at_shop = cooling_law(initial_temp_mix_at_shop, times)
len(temperatures_mix_at_shop)
```

###Now we find out the rate of cooling if we wait to add the cream back at the office.

```python
temperatures_unmixed_coffee = cooling_law(temp_coffee, times)
temperatures_mix_at_office = temp_mixture(temperatures_unmixed_coffee, vol_coffee, temp_cream, vol_cream)
len(temperatures_mix_at_office)
```

 ###Let's plot the two arrays

```python
plot(times, temperatures_mix_at_shop, 'o')
plot(times, temperatures_mix_at_office, 'D', color='r')
title('Coffee temperatures for different walking times')
xlabel('Walking time in minutes')
ylabel('Temperature in Fahrenheit')
legend(['Add cream at shop', 'Add cream at office'])
```

###From this graph it appears that it is better to mix the cream at the coffee shop, but I think the graph is a little ridiculous.  The red arrows are not real data points, the coffee was not that temperature at those times.

###It is clear, however, from comparing the final values of the arrays that the coffee is slightly cooler in the 20th minute if we  add the cream at the office.

###Maybe we want a more general purpose function where we can change the outside temperature.

```python
temp_coffee = 185.
vol_coffee = 8.
vol_cream = 1.
walk_time = 10.
r = 0.9

def temperature_difference(temp_cream, temp_out):
    temp_start = temp_mixture(temp_coffee, vol_coffee, temp_cream, vol_cream)
    temp_mix_at_shop = temp_out + (temp_start - temp_out) * r ** walk_time
    temp_start = temp_coffee
    temp_unmixed = temp_out + (temp_start - temp_out) * r ** walk_time
    temp_mix_at_office = temp_mixture(temp_unmixed, vol_coffee, temp_cream, vol_cream)
    return temp_mix_at_shop - temp_mix_at_office
```

###Let's add some different cream temperatures and outside temperatures and create a matrix of these values.

```python
cream_temperatures = arange(40., 51., 1.)
outside_temperatures = arange(35., 60., 1.)
cream_values, outside_values = meshgrid(cream_temperatures, outside_temperatures)
temperature_differences = temperature_difference(cream_values, outside_values)
```

###Create a  3D plot of the temperature differences.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
fig.set_size_inches(7,5)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(cream_values, outside_values, temperature_differences, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0)
xlabel('Cream temperature')
ylabel('Outside temperature')
title('Temperature difference')
```

###So that is fun, but what does it mean?  
###When considering the effect that the cream and the outside temperature have on the temperature of the coffee, we need to look at the relationship between the cream temperature and the outside temperature.
###This graph indicates that if the cream temperature and the outside temperature ate equal, it doesn't matter if we pour the cream into the coffee at the shop or the office, the final temperature at the office will be the same.
###However, if the cream is colder than the outside temperature (i.e. the blue area on the chart), you should mix it at the shop.
